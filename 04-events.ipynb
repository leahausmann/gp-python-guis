# Events & Callback-Funktionen

Tkinter reagiert auf **Ereignisse** (Events) mit **Callback-Funktionen**:
- Mausklick
- Tastatur

## Button-Callbacks über `command`


```python
import tkinter as tk

root = tk.Tk()
root.title('Events: Button-Callback')
root.geometry('250x120')

label = tk.Label(root, text='Noch kein Klick')
label.pack(pady=10)

def on_button_click():
    label.config(text='Button wurde geklickt!')

tk.Button(root, text='Klick mich', command=on_button_click).pack(pady=5)

root.mainloop()
```

## .bind()
Mit `.bind()` können wir auf Tastatur oder Maus reagieren

```python
import tkinter as tk

root = tk.Tk()
root.title('Events mit bind')
root.geometry('300x150')

label = tk.Label(root, text='Drücke eine Taste...')
label.pack(pady=20)

def on_key(event):
    label.config(text=f'Taste gedrückt: {event.keysym}')

root.bind('<Key>', on_key)

root.mainloop()
```
