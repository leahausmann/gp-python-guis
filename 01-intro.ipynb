## Was ist ein GUI?
GUI steht für **Graphical User Interface**.
Wir können eine grafische Benutzeroberfläche erstellen, und unsere Programme über Fenster, Buttons, Eingabefelder etc. bedienen

In Python können wir GUIs z.B mit **Tkinter** bauen. Tkinter ist das standardmäßige GUI-Framework von Python, wir müssen also nichts installieren.

## Erstes Tkinter-Fenster
Führe den folgenden Code in VS Code aus

```python
import tkinter as tk

# Hauptfenster erstellen
root = tk.Tk()
root.title('Mein erstes Tkinter-Fenster')
root.geometry('300x150')  # Breite x Höhe

label = tk.Label(root, text='Hello World!')
label.pack(pady=20)

# Event-Schleife starten
root.mainloop()
```
