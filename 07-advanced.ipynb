# Advanced: Messagebox, FileDialog, Canvas

fortgeschrittene Tkinter-Features:
- Messagebox (Nachrichtenfenster)
- FileDialog (Datei öffnen/speichern)
- Canvas (Zeichenfläche)

## Messagebox


```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title('Messagebox-Beispiel')
root.geometry('250x120')

def show_info():
    messagebox.showinfo('Info', 'Das ist eine Info-Messagebox.')

tk.Button(root, text='Info anzeigen', command=show_info).pack(pady=20)

root.mainloop()
```

## FileDialog – Datei öffnen


```python
import tkinter as tk
from tkinter import filedialog

root = tk.Tk()
root.title('FileDialog-Beispiel')
root.geometry('350x150')

label_file = tk.Label(root, text='Keine Datei gewählt')
label_file.pack(pady=10)

def choose_file():
    filepath = filedialog.askopenfilename(title='Datei auswählen')
    if filepath:
        label_file.config(text=f'Ausgewählte Datei:\n{filepath}')
    else:
        label_file.config(text='Keine Datei gewählt')

tk.Button(root, text='Datei auswählen', command=choose_file).pack(pady=5)

root.mainloop()
```

## Canvas – einfache Zeichenfläche


```python
import tkinter as tk

root = tk.Tk()
root.title('Canvas-Beispiel')
root.geometry('300x250')

canvas = tk.Canvas(root, bg='white')
canvas.pack(fill='both', expand=True)

# Ein paar Formen zeichnen
canvas.create_rectangle(20, 20, 120, 80, fill='lightblue', outline='blue')
canvas.create_oval(150, 50, 250, 150, fill='lightgreen', outline='green')
canvas.create_line(0, 0, 300, 200, fill='red', width=2)

root.mainloop()
```
