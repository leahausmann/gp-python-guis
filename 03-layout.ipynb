# Layouts: pack, grid, place

Tkinter hat drei Layout-Optionen:
- pack
- grid
- place

## pack()
pack() ordnet Widgets unter- oder nebeneinander an

```python
import tkinter as tk

root = tk.Tk()
root.title('Layout mit pack')
root.geometry('300x200')

tk.Label(root, text='Oben').pack(side='top', pady=5)
tk.Button(root, text='Links').pack(side='left', padx=5)
tk.Button(root, text='Rechts').pack(side='right', padx=5)
tk.Label(root, text='Unten').pack(side='bottom', pady=5)

root.mainloop()
```

## Tabellen
mit grid() kann man eine Tabelle machen

```python
import tkinter as tk

root = tk.Tk()
root.title('Layout mit grid')
root.geometry('250x120')

tk.Label(root, text='Benutzername:').grid(row=0, column=0, padx=5, pady=5, sticky='e')
entry_user = tk.Entry(root)
entry_user.grid(row=0, column=1, padx=5, pady=5)

tk.Label(root, text='Passwort:').grid(row=1, column=0, padx=5, pady=5, sticky='e')
entry_pw = tk.Entry(root, show='*')
entry_pw.grid(row=1, column=1, padx=5, pady=5)

tk.Button(root, text='Login').grid(row=2, column=0, columnspan=2, pady=10)

root.mainloop()
```

## genaue Koordinaten
place() positioniert Widgets mit Koordinaten (x, y).

```python
import tkinter as tk

root = tk.Tk()
root.title('Layout mit place')
root.geometry('300x200')

button1 = tk.Button(root, text='Button 1')
button1.place(x=30, y=30)

button2 = tk.Button(root, text='Button 2')
button2.place(x=150, y=100)

root.mainloop()
```
