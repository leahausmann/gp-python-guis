# Mini-App

Mini-App mit Tkinter: **Temperatur-Rechner**

## Ziel der App
- Eingabe in Celsius
- Button Umrechnen
- Ausgabe in Fahrenheit

Optional:
- Buttons für Richtung (C→F oder F→C)
- Fehlermeldung bei ungültiger Eingabe.

```python
import tkinter as tk

root = tk.Tk()
root.title('Temperatur-Converter')
root.geometry('320x200')

var_mode = tk.StringVar(value='C2F')

tk.Label(root, text='Temperatur eingeben:').pack(pady=5)
entry_value = tk.Entry(root)
entry_value.pack(pady=5)

frame_mode = tk.Frame(root)
frame_mode.pack(pady=5)
tk.Radiobutton(frame_mode, text='Celsius → Fahrenheit', variable=var_mode, value='C2F').pack(side='left')
tk.Radiobutton(frame_mode, text='Fahrenheit → Celsius', variable=var_mode, value='F2C').pack(side='left')

label_result = tk.Label(root, text='Ergebnis:')
label_result.pack(pady=10)

def convert():
    try:
        val = float(entry_value.get())
    except ValueError:
        label_result.config(text='Ergebnis: Ungültige Eingabe')
        return
    mode = var_mode.get()
    if mode == 'C2F':
        res = val * 9/5 + 32
        label_result.config(text=f'Ergebnis: {res:.2f} °F')
    else:
        res = (val - 32) * 5/9
        label_result.config(text=f'Ergebnis: {res:.2f} °C')

tk.Button(root, text='Umrechnen', command=convert).pack(pady=5)

root.mainloop()
```
