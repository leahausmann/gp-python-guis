# Widgets: Buttons, Labels, Entry

In diesem Notebook lernst du die wichtigsten Basis-Widgets kennen:
- Label (Text)
- Button (Button)
- Entry (Eingabe)

## Label & Button


```python
import tkinter as tk

root = tk.Tk()
root.title('Widgets: Label & Button')
root.geometry('300x150')

label = tk.Label(root, text='Klicke auf den Button')
label.pack(pady=10)

def on_click():
    label.config(text='Button wurde geklickt!')

button = tk.Button(root, text='Klick mich', command=on_click)
button.pack(pady=5)

root.mainloop()
```

## Entry (Eingabefeld)
Mit Entry kann der Benutzer Text eingeben.

```python
import tkinter as tk

root = tk.Tk()
root.title('Widget: Entry')
root.geometry('320x180')

tk.Label(root, text='Wie heißt du?').pack(pady=5)
entry_name = tk.Entry(root)
entry_name.pack(pady=5)

label_output = tk.Label(root, text='')
label_output.pack(pady=10)

def greet():
    name = entry_name.get().strip()
    if name:
        label_output.config(text=f'Hallo, {name}!')
    else:
        label_output.config(text='Bitte gib einen Namen ein.')

tk.Button(root, text='Begrüßen', command=greet).pack(pady=5)

root.mainloop()
```
